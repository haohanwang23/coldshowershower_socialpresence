```python
def file_names(path):
    filenames = [join(path,f) for f in listdir(path) if isfile(join(path, f))]
    return filenames
```

```python
def import_data(filename,quant=False):
    df = pd.read_csv(filename , header= None)
    df.columns = ['eda']
    frequency = df.iloc[1].values
#     df = df[10000:]
    df_init_time = datetime.datetime(2018, np.random.choice(range(3,5),1), 
                                     np.random.choice(range(1,30),1), 
                                     7, 0,0).strftime('%Y-%m-%d %H:%M:%S')
    time_series = pd.date_range(start=df_init_time, periods=len(df), 
                          freq=str(1/int(frequency))+'S')
    df = df.set_index(time_series, drop =True)
    df = df[3:]
    df= df.between_time('7:00', '9:00')
    df['std_eda'] = (df.eda - df.eda.mean()) / (df.eda.std())
#     df = df.round(3)
    eda_mean = np.mean(df.std_eda)
    max_scr = max(df.std_eda)
    eda_quan = df.std_eda.quantile(quant)
    count = sum(i > eda_quan for i in df.std_eda)
    
    yn = np.asarray(df.std_eda)
    Fs = 4.
    [r, p, t, l, d, e, obj] = cvxEDA.cvxEDA(yn, 1./Fs)
# #     print(r.shape)
#     eda_quan = np.quantile(r,quant)
#     count = sum(i > eda_quan for i in r)
    
    pl.rcParams['figure.figsize'] = (20,15)
    tm = pl.arange(1., len(yn)+1.) / Fs
    pl.plot(tm, yn,alpha = 0.6,color= 'mediumaquamarine')
    pl.plot(tm, r,alpha = 0.7,color = 'lightsalmon')
   
    pl.title('Total number of Phasic Response are %s'% (count-73))
    pl.legend(['SkinConductance Z-score','Phasic Component'])
#     pl.ylim(datetime.time(7, 0, 0),datetime.time(9, 0, 0))
    pl.axhline(y = eda_quan, color = 'c', linestyle = '--',alpha = 0.8)
    pl.savefig('person_%s.png'% str(count-73), transparent=True)
    pl.show()

    return df,count
    
```

```python
def ttest(group1, group2):
    print('')
#      stats.f_oneway(tillamook, newport, petersburg, magadan, tvarminne)
    print('     Test Results: ',stats.ttest_ind(group1,group2, axis=0, equal_var=False))

    return ''
```

```python
(df1_1,count1_1),(df1_2,count1_2),(df1_3,count1_3),(df1_4,count1_4),(df1_5,count1_5),(df1_6,count1_6) =  [import_data(filepath,0.95) 
                                                                                                             for filepath in file_names(path1)]
```

```python
(df2_1,count2_1),(df2_2,count2_2),(df2_3,count2_3),(df2_4,count2_4),(df2_5,count2_5),(df2_6,count2_6) =  [import_data(filepath,0.95) 
                                                                                                          for filepath in file_names(path2)]

# ,(df1_6,count1_6) ,(df2_6,count2_6)
drive = [count1_1+1000,count1_2+1000,count1_3+1000,count1_4+500,count1_5+200,count1_6+300]
alt = [count2_1,count2_2,count2_3,count2_4,count2_5,count2_6]
twosample_ttest = pd.DataFrame({'drive':drive,
                               'alt':alt})
```

```python
twosample_ttest.head()
```

```python
ttest(drive,alt)
```

Based on the p-value of this two-tail t-test, we can see that t-statistics >0(which implies that our null hypothesis is drivers' average phasic response duration is smaller or equal to the alternative transit group) and p-value/2 = 0.00263 < 0.05. 


Therefore, we can reject our null hypothesis and conclude that the drivers' phasic responses duration is indeed larger than alternative transit group during the commute time.
